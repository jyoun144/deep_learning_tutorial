$\begin{equation*}
\mathbf{z}_{h_1} = \mathbf{w}_1 \times \mathbf{x}_1 + \mathbf{w}_2 \times \mathbf{x}_2 + \mathbf{b}_1 \\
\mathbf{z}_{h_2} = \mathbf{w}_3 \times \mathbf{x}_1 + \mathbf{w}_4 \times \mathbf{x}_2 + \mathbf{b}_1 \\
\mathbf{a}_{h_1} = \frac{1}{1 + e^{-\mathbf{z}_{h_1}}} \\
\mathbf{a}_{h_2} = \frac{1}{1 + e^{-\mathbf{z}_{h_2}}} \\ 
\mathbf{z}_{o_1} = \mathbf{w}_5 \times \mathbf{a}_{h_1} + \mathbf{w}_6 \times \mathbf{a}_{h_2} + \mathbf{b}_2 \\
\mathbf{z}_{o_2} = \mathbf{w}_7 \times \mathbf{a}_{h_1} + \mathbf{w}_8 \times \mathbf{a}_{h_2} + \mathbf{b}_2 \\
\mathbf{a}_{o_1} = \frac{1}{1 + e^{-\mathbf{z}_{o_1}}} \\
\mathbf{a}_{o_2} = \frac{1}{1 + e^{-\mathbf{z}_{o_2}}} \\
\mathbf{Error}_1 = \frac{1}{2}\Bigl(\mathbf{target}_{a_{o_1}}-\mathbf{a}_{o_1}\Bigl)^2 \\
\mathbf{Error}_2 = \frac{1}{2}\Bigl(\mathbf{target}_{a_{o_2}} - \mathbf{a}_{o_2}\Bigl)^2 \\
\mathbf{Total}_{Error} = \mathbf{Error}_1  + \mathbf{Error}_2 
\end{equation*}$

```python
import math

class simple_neural_network:
    def __init__(self, model_params, num_of_epochs):        
        self.model_params = model_params
        self.num_of_epochs = num_of_epochs    
        
    def train_model(self):
        activations_cache = []
        current_params = self.model_params.copy()
        for i in range(self.num_of_epochs):
            activations_hidden_layer = self._forward_propagation_hidden_layer(current_params)
            activations_output_layer = self._forward_propagation_output_layer(current_params, activations_hidden_layer)
            activations_cache.append({'activations_hidden_layer':activations_hidden_layer,
                                     'activations_output_layer':activations_output_layer})
        return activations_cache
    
    def _forward_propagation_hidden_layer(self, params):
        z_h1 = params['w1'] * params['i1'] + params['w2'] * params['i2']  + params['b1'][0] * params['b1'][1]
        z_h2 = params['w3'] * params['i1'] + params['w4'] * params['i2']  + params['b1'][0] * params['b1'][1]
        a_h1 = self._sigmoid(z_h1)
        a_h2 = self._sigmoid(z_h2)
        return {'z_h1': z_h1, 'z_h2':z_h2, 'a_h1': a_h1, 'a_h2': a_h2 }
    
    def _forward_propagation_output_layer(self, params, activations):
        z_o1 = params['w5'] * activations['a_h1'] + params['w6'] * activations['a_h2']  + params['b2'][0] * params['b2'][1]
        z_o2 = params['w7'] * activations['a_h1'] + params['w8'] * activations['a_h2']  + params['b2'][0] * params['b2'][1]
        a_o1 = self._sigmoid(z_o1)
        a_o2 = self._sigmoid(z_o2)
        return {'z_o1': z_o1, 'z_o2':z_o2, 'a_o1': a_o1, 'a_o2': a_o2 }
            
    def _sigmoid(self, z_input):
        a_ouput = 1/(1 + math.exp(-1.0 * z_input))
        return a_ouput
```

```python
input_dict = {'i1': 0.05,
              'i2': 0.10,
              'w1': 0.15,
              'w2': 0.20,
              'w3': 0.25,
              'w4': 0.30,
              'w5': 0.40,
              'w6': 0.45,
              'w7': 0.50,
              'w8': 0.55,
              'b1': (1.00, 0.35),
              'b2': (1.00, 0.60)}
obj = simple_neural_network(input_dict, 1)
result = obj.train_model()
for item in result:
    for key1 in item:
        for key2 in item[key1]:
            print('{}: {}\n'.format(key2, item[key1][key2]))
```
